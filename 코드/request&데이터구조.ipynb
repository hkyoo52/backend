### 아이피주소 크롤링하기
* 실제로 서버는 220.76.23.130만 제공
* 브라우저가 Elements 형식으로 바꿈

```python
# https://api.ipify.org/ 에서 알 수 있음
# 220.76.23.130

import requests as req

res = req.get('https://api.ipify.org/')
print(res.status_code) # 브라우저 요청에 대한 서버 반은
print(res.text)

```

```python
res = req.get('https://api.ipify.org/')
print(res.request.method)
```

```python
res = req.get('https://api.ipify.org/')
print(res.request.headers)
```

```python
res = req.get('https://api.ipify.org/')
print(res.raw)
```
